## Get Daily Token

```python
import json
import os

import httpx
from dotenv import find_dotenv, load_dotenv

URL = "https://eservice.ura.gov.sg/uraDataService/insertNewToken/v1"

load_dotenv(find_dotenv(usecwd=True))
access_key = os.environ.get("URA_KEY")
if not access_key:
    raise RuntimeError(
        "URA_KEY is not set. Export it or add URA_KEY=... to the project .env file."
    )

headers = {
    "AccessKey": access_key,
    "Accept": "application/json",
    "User-Agent": "curl/8.7.1",
}

try:
    with httpx.Client(timeout=30.0) as client:
        resp = client.get(URL, headers=headers)
        resp.raise_for_status()
        raw = resp.text
except httpx.HTTPStatusError as e:
    print(e.response.status_code, e.response.reason_phrase)
    print(e.response.text)
    raise

try:
    data = json.loads(raw)
    print(json.dumps(data, indent=2))
    if isinstance(data, dict) and data.get("Status") == "Success":
        token = data.get("Result", "")
        print("\nToken (paste into .env as URA_TOKEN):", token)
except json.JSONDecodeError:
    print(raw)
```
